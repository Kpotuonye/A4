Adding layers to the map

```python
# Create a Map instance
m = folium.Map(location=[47.8095, 13.0550],
    zoom_start=12, control_scale=True)

# Add marker
# Run: help(folium.Icon) for more info about icons
folium.Marker(
    location=[47.8095, 13.0550],
    popup='Salzburg city',
    icon=folium.Icon(color='green', icon='ok-sign'),
).add_to(m)

#Show map
```

Loading further data

```python
# File path
points_fp = r"C:\Users\Tarzan\Downloads\museum.shp"

# Read the data
points = gpd.read_file(points_fp)


#Check input data
points.head()
```

```python
points.head()
```

converting points to GeoJSON features using folium

```python
points_gjson = folium.features.GeoJson(points, name="Salzburg Museums")
```

```python
#Adding points to basemap
# Create a Map instance
m = folium.Map(location=[47.8095, 13.0550], tiles = 'cartodbpositron', zoom_start=11, control_scale=True)

# Add points to the map instance
points_gjson.add_to(m)

#Show map
m
```

We add a LayerControl object on our map, which allows the tourist control which map layers are visible

```python
# Create a layer control object and add it to our map instance
folium.LayerControl().add_to(m)

#Show map
m
```

Heatmap is one of the folium plugins which creates heatmap layer from input points.

We now visualize a heatmap of the major museums in Salzburg using the addresses input data

```python
# Create a list of coordinate pairs
#locations = list(zip((47.80344, 13.04191),(47.79822, 13.04299),(47.79798, 13.04775),(47.79525, 13.04732),(47.79983, 13.03996),(47.7621, 13.0603),(47.7985, 13.0457),(47.7973, 13.0445)))
#x = points["geometry"].x
#points["y"] = points["geometry"].y
toPoints= points["geometry"].explode(index_parts=True)
display(toPoints.head())
x = [point.x for point in toPoints]
y = [point.y for point in toPoints]
```

```python
positions = list(zip(y, x))
print(positions)
```

```python
from folium.plugins import HeatMap

# Create a Map instance
m = folium.Map(location=[47.8095, 13.0550], tiles = 'stamentoner', zoom_start=10, control_scale=True)

# Add heatmap to map instance
# Available parameters: HeatMap(data, name=None, min_opacity=5, max_zoom=18, max_val=1.0, radius=50, blur=30, gradient=None, overlay=True, control=True, show=True)
HeatMap(positions).add_to(m)

# Show map
m
```
